```python
import pandas as pd
import numpy as np
# Faz alteração no estilo do DataFrame
import seaborn as sns
```

```python
# Criando um DataFrame
df = pd.DataFrame( index = ['Carol', 'Ana', 'Virginia', 'Bruna', 'Beatriz', 'Breno'],
                  data = [[8.9,9.9,8.8,6.6, 6., 7.],[6.6, 7.5, 7.6, 6.5, 8.5, np.nan],[6.4, 9., 6.6,7.5,8.6, 9.],
                          [5.5, 7.4,6.7,9.9,7., np.nan],[9., 6.6,7.,9.9,10., 8.],[5.5,7.7,8.9,5.5,6.9, 10.]],
                  columns = ['Potugues', 'Quimica', 'Fisica', 'Matematica', 'Historia', 'Geografia'])
df_ = sns.light_palette("green", as_cmap=True)

s = df.style.background_gradient(cmap=df_)
s
```

```python
# Visualiza no incio ao fim
df[:].style.background_gradient(cmap=df_)
```

```python
# Visualiza do 0 ao 3
df[0:3].style.background_gradient(cmap=df_)
```
